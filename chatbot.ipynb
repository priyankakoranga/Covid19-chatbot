```python
import json
import numpy as np
import tensorflow as tf
from tensorflow import keras
import nltk
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
with open('intents.json') as file:
    data=json.load(file)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
words=[]
labels=[]
training_sentences=[]
training_labels=[]


for intent in data['intents']:
    for pattern in intent['patterns']:
        training_sentences.append(pattern)
        training_labels.append(intent['tag'])
   # responses.append(intent['responses'])
    
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
```

```python
enc= LabelEncoder()
enc.fit(training_labels)
training_labels=enc.transform(training_labels)
```

```python
vocab_size=1000
embedding_dim=16
max_len=20

tokenizer=Tokenizer(num_words=vocab_size,oov_token='<OOV>') #oov is used to add token that is out of vocabulary
tokenizer.fit_on_texts(training_sentences)
word_index= tokenizer.word_index

sequences=tokenizer.texts_to_sequences(training_sentences)

padded=pad_sequences(sequences,truncating='post',maxlen=max_len)
```

```python
clases=len(labels)
```

```python
model=tf.keras.models.Sequential()
model.add(keras.layers.Embedding(vocab_size,embedding_dim,input_length=max_len))
model.add(keras.layers.GlobalAveragePooling1D())
model.add(keras.layers.Dense(16,activation='relu'))
model.add(keras.layers.Dense(16,activation='relu'))
model.add(keras.layers.Dense(clases,activation='softmax'))
```

```python
model.summary()
```

```python
training_labels_final=np.array(training_labels)
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(padded,training_labels_final,epochs=200)
```

```python
def check():
    print('start talking with bot, Enter quit to exit')
    while True:
        string= input('Enter: ')
        if string == 'quit': break
        result =model.predict(pad_sequences(tokenizer.texts_to_sequences([string]),truncating='post',maxlen=max_len))[0]                   
        category = enc.inverse_transform([np.argmax(result)]) 
        
        for i in data['intents']:
            if i['tag']==category:
                print(np.random.choice(i['responses']))
        
```

```python
check()
```
